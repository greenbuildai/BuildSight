```python
import cv2
import numpy as np
points = []

def draw_polygon(event, x, y, flags, param):
    global points
    if event == cv2.EVENT_LBUTTONDOWN:
        points.append((x, y))

cap = cv2.VideoCapture("rd1.mp4")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    for p in points:
        cv2.circle(frame, p, 5, (0, 0, 255), -1)

    if len(points) > 1:
        cv2.polylines(frame, [np.array(points)], False, (0, 0, 255), 2)

    cv2.imshow("Draw Danger Zone - Press Q when done", frame)
    cv2.setMouseCallback("Draw Danger Zone - Press Q when done", draw_polygon)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

print("DANGER ZONE POINTS:", points)

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# =====================================
# DEFINE DANGER ZONE (EDIT THESE POINTS)
# =====================================
DANGER_ZONE = np.array([(421, 219), (672, 199), (586, 58), (430, 61), (43, 107), (4, 240), (420, 221)], dtype=np.int32)

# =====================================
# LOAD VIDEO
# =====================================
VIDEO_PATH = "rd1.mp4"
cap = cv2.VideoCapture(VIDEO_PATH)

if not cap.isOpened():
    print("❌ Error: Cannot open video")
    exit()

print("✅ Video loaded successfully")

# =====================================
# VIDEO LOOP
# =====================================
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Draw polygon outline
    cv2.polylines(
        frame,
        [DANGER_ZONE],
        isClosed=True,
        color=(0, 0, 255),   # Red
        thickness=2
    )

    # Transparent fill
    overlay = frame.copy()
    cv2.fillPoly(overlay, [DANGER_ZONE], (0, 0, 255))
    frame = cv2.addWeighted(overlay, 0.25, frame, 0.75, 0)

    # Label
    cv2.putText(
        frame,
        "DANGER ZONE",
        (DANGER_ZONE[0][0], DANGER_ZONE[0][1] - 10),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.8,
        (0, 0, 255),
        2
    )

    cv2.imshow("Danger Zone Visualization", frame)

    # Press Q to quit
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

# =====================================
# CLEANUP
# =====================================
cap.release()
cv2.destroyAllWindows()
```
