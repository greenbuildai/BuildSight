```python
import cv2
import numpy as np

# =====================================
# DEFINE DANGER ZONE (EDIT THESE POINTS)
# =====================================
DANGER_ZONE = np.array([(258, 227), (174, 447), (413, 465), (426, 222), (256, 225)], dtype=np.int32)

# =====================================
# LOAD VIDEO
# =====================================
VIDEO_PATH = "rd1.mp4"
cap = cv2.VideoCapture(VIDEO_PATH)

if not cap.isOpened():
    print("❌ Error: Cannot open video")
    exit()

print("✅ Video loaded successfully")

# =====================================
# VIDEO LOOP
# =====================================
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Draw polygon outline
    cv2.polylines(
        frame,
        [DANGER_ZONE],
        isClosed=True,
        color=(0, 0, 255),   # Red
        thickness=2
    )

    # Transparent fill
    overlay = frame.copy()
    cv2.fillPoly(overlay, [DANGER_ZONE], (0, 0, 255))
    frame = cv2.addWeighted(overlay, 0.25, frame, 0.75, 0)

    # Label
    cv2.putText(
        frame,
        "DANGER ZONE",
        (DANGER_ZONE[0][0], DANGER_ZONE[0][1] - 10),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.8,
        (0, 0, 255),
        2
    )

    cv2.imshow("Danger Zone Visualization", frame)

    # Press Q to quit
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

# =====================================
# CLEANUP
# =====================================
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

DANGER_ZONE = np.array([
    (258, 227),
    (174, 447),
    (413, 465),
    (426, 222)
], dtype=np.int32)

VIDEO_PATH = r"E:\Company\Green Build AI\Prototypes\BuildSight\buildsight-base\gis\inputs\rd1.mp4"

cap = cv2.VideoCapture(VIDEO_PATH)

if not cap.isOpened():
    print("❌ Error opening video")
    exit()

while True:

    ret, frame = cap.read()

    if not ret or frame is None:
        break

    frame = cv2.resize(frame, (960, 540))

    overlay = frame.copy()

    cv2.polylines(frame, [DANGER_ZONE], True, (0,0,255), 2)
    cv2.fillPoly(overlay, [DANGER_ZONE], (0,0,255))

    frame = cv2.addWeighted(overlay, 0.25, frame, 0.75, 0)

    cv2.imshow("Danger Zone", frame)

    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
